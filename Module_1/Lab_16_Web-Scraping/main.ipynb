#### Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
#PRINTING THE POPULAR REPOS

# your code here
html = requests.get(url).content
html_soup = BeautifulSoup(html, "html.parser")

# Find all tags <a> with "class" : "h3 lh-condensed" & "class : "h4 lh-condensed"
dev_tags = html_soup.find_all("h1", {"class" : "h3 lh-condensed"})
dev_repos_tags = html_soup.find_all("h1", {"class" : "h4 lh-condensed"})

for dev_tag, repo_tag in zip(dev_tags, dev_repos_tags):
    dev_name = dev_tag.find("a").text.strip()
    repo_name = repo_tag.find("a").text.strip()
    print(f"{dev_name} ({repo_name})")
````

#### Display all the image links from Walt Disney wikipedia page.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
#DISPLAYING ALL THE IMAGE LINKS FROM THE WALT DISNEY WIKI PAGE

# your code here
html = requests.get(url).content
html_soup = BeautifulSoup(html, "html.parser")


# Find all tags <a> with "class" : "image" 
images_tags = html_soup.find_all("a", {"class" : "image"})

for image_tag in images_tags:
    page_image = image_tag.find("img")
    img_url = page_image["src"]
    print(img_url)
    
    


# # boucler sur chaque balise <a> trouvée
# for image_tag in images_tags:
#     # extraire la balise <img> à l'intérieur de la balise <a>
#     img_tag = image_tag.find("img")
#     # extraire l'URL de l'image à partir de l'attribut "src" de la balise <img>
#     img_url = img_tag["src"]
#     # afficher l'URL de l'image
#     print(img_url)

````
