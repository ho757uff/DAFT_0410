```python
result = response.json()
```

```python
import pprint

pprint.pprint(result)
```

```python
repo_files_names = [item["name"] for item in result["items"]]
repo_files_names.sort()
repo_files_names
```

```python
repo_files_urls = {item["name"]: item["url"] for item in result["items"]}
repo_files_urls
```

```python
files = [req.get(repo_files_urls[repo_file_name]) for repo_file_name in repo_files_names]
files
```

```python
repo_files_content = [file.json()["content"] for file in files]
repo_files_content
```

```python
import base64

message = [base64.b64decode(b64message).decode('ascii').replace("\n","") for b64message in repo_files_content]
message
```
