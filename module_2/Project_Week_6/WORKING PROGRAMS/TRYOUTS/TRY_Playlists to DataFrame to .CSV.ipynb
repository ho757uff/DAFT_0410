## Exploring the list of songs inside the playlists

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="FR")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

## Extracting songs from playlists

```python
list_of_audio_features1 = []
for item in range(0,1028):
    #print (tracks[item]["track"]["id"])
    list_of_audio_features1.append(sp.audio_features(tracks1[item]["track"]["id"])[0])
```

```python
list_of_audio_features2 = []
for item in range(0,5200):
    #print (tracks[item]["track"]["id"])
    list_of_audio_features2.append(sp.audio_features(tracks2[item]["track"]["id"])[0])
```

```python
list_of_audio_features3 = []
for item in range(0,1101):
    #print (tracks[item]["track"]["id"])
    list_of_audio_features3.append(sp.audio_features(tracks3[item]["track"]["id"])[0])
```

```python
list_of_audio_features4 = []
for item in range(0,3595):
    #print (tracks[item]["track"]["id"])
    list_of_audio_features4.append(sp.audio_features(tracks4[item]["track"]["id"])[0])
```
